# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# create variables for each type of city
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

rural_cities_df
```

```python
#  1. Get the total rides for each city type
urban_ride_count = urban_cities_df.count()["ride_id"]
suburban_ride_count = suburban_cities_df.count()["ride_id"]
rural_ride_count = rural_cities_df.count()["ride_id"]

rural_ride_count
```

```python
# 2. Get the total drivers for each city type
urban_driver_count = city_data_df[city_data_df["type"] == "Urban"].sum()["driver_count"]
suburban_driver_count = city_data_df[city_data_df["type"] == "Suburban"].sum()["driver_count"]
rural_driver_count = city_data_df[city_data_df["type"] == "Rural"].sum()["driver_count"]


rural_driver_count
```

```python
#  3. Get the total amount of fares for each city type
urban_fare_count = urban_cities_df.sum()["fare"]
suburban_fare_count = suburban_cities_df.sum()["fare"]
rural_fare_count = rural_cities_df.sum()["fare"]

urban_fare_count 
```

```python
#  4. Get the average fare per ride for each city type. 
urban_avg_fare = urban_fare_count / urban_ride_count
suburban_avg_fare = suburban_fare_count / suburban_ride_count
rural_avg_fare = rural_fare_count / rural_ride_count

urban_avg_fare
```

```python
# 5. Get the average fare per driver for each city type. 
urban_driver_avg_fare = urban_fare_count / urban_driver_count
suburban_driver_avg_fare = suburban_fare_count / suburban_driver_count
rural_driver_avg_fare = rural_fare_count / rural_driver_count

urban_driver_avg_fare
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": [rural_ride_count , suburban_ride_count , urban_ride_count],
    "Total Drivers": [rural_driver_count, suburban_driver_count , urban_driver_count],
    "Total Fares": [rural_fare_count, suburban_fare_count , urban_fare_count],
    "Average Fare per Ride": [rural_avg_fare, suburban_avg_fare ,urban_avg_fare],
    "Average Fare per Driver": [ rural_driver_avg_fare, suburban_driver_avg_fare ,urban_driver_avg_fare]},
index= ["Rural", "Suburban", "Urban"])
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None 
```

```python
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"]  = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

#  varible x = variable x .map("{:,.1f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_date = pyber_data_df.groupby(['type', 'date'])
sum_fares_date = fares_date.sum()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_date = sum_fares_date.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_summary_pivot = sum_fares_date.pivot(index= 'date',
       columns= 'type',
       values= 'fare')
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
Jan_Apr_df = pyber_summary_pivot.loc['2019-01-01':'2019-04-29']
Jan_Apr_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
Jan_Apr_df.index = pd.to_datetime(Jan_Apr_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
Jan_Apr_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = Jan_Apr_df.resample('W').sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

weekly_fares_df = weekly_fares_df.plot.line(figsize=(20,6))
# Add a title 
weekly_fares_df.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
weekly_fares_df.set_xlabel("")
weekly_fares_df.set_ylabel("Fare ($USD)")


plt.show()
plt.tight_layout()

plt.savefig('analysis/PyBer_fare_summary.png')
```
